Según los datos disponibles, hay **21 212** (veintiún mil doscientos doce) apellidos diferentes registrados en el país.

En los archivos, cada columna representa:

- **Apellido**: *primer apellido* de la persona empadronada como votante en esa provincia. En el formato de los documentos provistos están en letras mayúsculas, sin tildes pero sí con Ñ. Ejemplo: NUÑEZ.
- **Personas**: *cantidad de votantes* empadronados con ese apellido en esa provincia.
- **Porcentaje**: *porcentaje del padrón provincial* que representa ese apellido. En el caso de la información para Costa Rica en general, este es el porcentaje del padrón nacional.

### Definiciones

A partir de las *frecuencias relativas* de los apellidos (la cantidad de personas que tienen ese primer apellido en una provincia en comparación con el total de personas registradas en esa provincia) es posible establecer algunas relaciones de probabilidad, como las siguientes:

- $P(X)$ es la probabilidad de tener el apellido $X$ a nivel nacional.
- $P(P_i)$ es la probabilidad de estar empadronado en la provincia $P_i$.
- $P(X \mid P_i)$ es la probabilidad de tener el apellido $X$ si estoy empadronado en la provincia $P_i$.
- $P(P_i \mid X)$ es la probabilidad de estar empadronado en la provincia $P_i$ si tengo el apellido $X$.

Y la siguiente relación se cumple:

\begin{equation}
\begin{aligned}
P(X \cap P_i) & = P(P_i \cap X) \\
P(X) P(P_i \mid X) & = P(P_i) P(X \mid P_i) \\
\end{aligned}
\end{equation}

Podríamos estar interesados en conocer, por ejemplo, cuál es la probabilidad de que una persona elegida al azar esté empadronada en la provincia $P_i$ dado que su apellido es $X$, es decir, $P(P_i \mid X)$, entonces despejamos para:

\begin{equation}
P(P_i \mid X) = \frac{P(P_i) P(X \mid P_i)}{P(X)}
\end{equation}

Que se convierte en la regla de la probabilidad condicional inversa (teorema de Bayes).
