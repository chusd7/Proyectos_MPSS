### Definiciones

A partir de las *frecuencias relativas* de los apellidos (la cantidad de personas que tienen ese primer apellido en una provincia en comparaci√≥n con el total de personas registradas en esa provincia) es posible establecer algunas relaciones de probabilidad, como las siguientes:

- $P(X)$ es la probabilidad de tener el apellido $X$ a nivel nacional.
- $P(P_i)$ es la probabilidad de estar empadronado en la provincia $P_i$.
- $P(X \mid P_i)$ es la probabilidad de tener el apellido $X$ si estoy empadronado en la provincia $P_i$.
- $P(P_i \mid X)$ es la probabilidad de estar empadronado en la provincia $P_i$ si tengo el apellido $X$.

Y la siguiente relaci√≥n se cumple:

\begin{equation}
\begin{aligned}
P(X \cap P_i) & = P(P_i \cap X) \\
P(X) P(P_i \mid X) & = P(P_i) P(X \mid P_i) \\
\end{aligned}
\end{equation}

Podr√≠amos estar interesados en conocer, por ejemplo, cu√°l es la probabilidad de que una persona elegida al azar est√© empadronada en la provincia $P_i$ dado que su apellido es $X$, es decir, $P(P_i \mid X)$, entonces despejamos para:

\begin{equation}
P(P_i \mid X) = \frac{P(P_i) P(X \mid P_i)}{P(X)}
\end{equation}

Que se convierte en la regla de la probabilidad condicional inversa (teorema de Bayes).
