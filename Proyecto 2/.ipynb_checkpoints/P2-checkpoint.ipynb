## Problema

El archivo `spotify.csv` tiene una selección de más de veinte mil canciones que representan las 10 más escuchadas de más de dos mil artistas distintos de la plataforma Spotify.

<img src="https://storage.googleapis.com/pr-newsroom-wp/1/2018/11/Spotify_Logo_RGB_Green.png" width="200px">

Cada una de estas canciones tiene la siguiente información:

- `artist`: nombre del artista.
- `track`: nombre de la canción tal como se muestra en la plataforma de Spotify.
- `album`: el álbum en el que se encuentra la canción en Spotify.
- `danceability`: describe qué tan adecuada es una pista para bailar según una combinación de elementos musicales que incluyen tempo, estabilidad del ritmo, fuerza del ritmo y regularidad general. Un valor de 0.0 es el menos adecuado para bailar y 1.0 es el más adecuado.
- `energy`: es una medida de 0.0 a 1.0 que representa una medida perceptual de intensidad y actividad. Las pistas energéticas suelen sentirse rápidas, fuertes y ruidosas. Por ejemplo, el *death metal* tiene una energía alta, mientras que un preludio de Bach tiene una puntuación baja en la escala. Las características perceptuales que contribuyen a este atributo incluyen el rango dinámico, la percepción de la sonoridad, el timbre, la velocidad de inicio y la entropía general.
- `key`: la clave en la que se encuentra la canción. Los enteros se asignan a notas usando la notación estándar de clases de tono. Por ejemplo, 0 = C (do), 1 = C♯/D♭ (do sostenido o re bemol), 2 = D (re), y así sucesivamente. Si no se detectó ninguna clave, el valor es -1.
- `loudness`: el volumen general de una pista en decibel (dB). Los valores de sonoridad se promedian en toda la pista y son útiles para comparar el volumen relativo de las pistas. La sonoridad es la cualidad de un sonido que es la correlación psicológica primaria de la fuerza física (amplitud). Los valores suelen oscilar entre -60 y 0 dB.
- `speechiness`: detecta la presencia de palabras habladas en una pista. Cuanto más parecida a la grabación exclusivamente hablada (por ejemplo, programa de entrevistas, audiolibro, poesía), más cercano a 1.0 es el valor del atributo. Los valores por encima de 0,66 describen pistas que probablemente estén hechas completamente de palabras habladas. Los valores entre 0,33 y 0,66 describen pistas que pueden contener tanto música como palabras habladas, ya sea en secciones o en capas, incluyendo casos como la música rap. Los valores por debajo de 0,33 probablemente representan música y otras pistas no parecidas al habla.
- `acousticness`: una medida de confianza de 0.0 a 1.0 de si la canción es acústica. 1.0 representa una alta confianza de que la canción es acústica.
- `instrumentalness`: predice si una pista no contiene vocales. Los sonidos "ooh" y "aah" se tratan como instrumentales en este contexto. Las pistas de rap o de palabra hablada son claramente "vocales". Cuanto más cercano sea el valor de *instrumentalness* a 1.0, mayor será la probabilidad de que la pista no contenga contenido vocal. Los valores por encima de 0.5 se destinan a representar pistas instrumentales, pero la confianza es mayor a medida que el valor se acerca a 1.0.
- `liveness`: detecta la presencia de una audiencia en la grabación. Los valores más altos de *liveness* representan una mayor probabilidad de que la pista se haya interpretado en vivo. Un valor por encima de 0.8 proporciona una alta probabilidad de que la pista sea en vivo.
- `valence`: una medida de 0.0 a 1.0 que describe la positividad musical transmitida por una pista. Las pistas con un *valence* alto suenan más positivas (por ejemplo, felices, alegres, eufóricas), mientras que las pistas con un *valence* bajo suenan más negativas (por ejemplo, tristes, deprimidas, enojadas).
- `tempo`: el *tempo* estimado general de una pista en beats por minuto (BPM). En la terminología musical, el *tempo* es la velocidad o el ritmo de una pieza dada y deriva directamente de la duración promedio del *beat* (BPM, beats por minuto).
- `duration_ms`: la duración de la pista en milisegundos.
- `streams`: número de streams de la canción en Spotify.
- `album_type`: indica si la canción se lanzó en Spotify como sencillo o si está incluida en un álbum o compilado.
- `url`: la URL de la canción.
- `uri`: un enlace de Spotify utilizado para encontrar la canción a través de la API.

**Nota 1**: Estos datos están adaptados de [Kaggle](https://www.kaggle.com/datasets/salvatorerastelli/spotify-and-youtube). Datos recopilados el 7 de febrero de 2023.

**Nota 2**: De estos parámetros, los 10 "***parámetros musicales***" son: *danceability*, *energy*, *key*, *loudness*, *speechiness*, *acousticness*, *instrumentalness*, *liveness*, *valence*, *tempo*.

1.(5 %) ¿Cuántos tipos de álbumes hay? ¿Cuántos álbumes diferentes hay?

````python
# Para saber cuantos tipos de álbumes diferentes hay, se usará
# La función nunique() que realiza una iteración en este caso
# De una columna para poder determinar cuantos álbumes
# Diferentes hay.
# Tipos de álbumes usando función nunique
num_tipo_album = songs['album_type'].nunique()
txt1 = 'La cantidad de tipos de álbumes es de: '
print(txt1 + str(num_tipo_album))
# Para saber cuantos álbumes diferentes hay se usa el mismo
# Código anterior cambiando la columna a 'album'
# Números de álbumes diferentes usando función nunique
# Se filtran las filas donde album_type sea igual a album y
# Luego se llama la función nunique para obtener los valores únicos
num_album = songs[songs['album_type'] == 'album']['album'].nunique()
# Se imprime el resultado
txt2 = 'La cantidad de álbumes diferentes de tipo album es de: '
print(txt2 + str(num_album))
````

2.(5 %) ¿Cuál es la canción más corta y más larga de la lista según los datos?

````python
# Para este caso se usarán las funciones idxmin() e idxmax() de
# La librería pandas,estas funciones permiten obtener el
# Valor numérico mínimo y máximo respectivamente de una
# Columna y devuelven el id de una columna seleccionada.
# Se obtiene la canción con la duración mas baja
cancion_duracion_min = songs.loc[songs['duration_ms'].idxmin(), 'track']
txt3 = 'El nombre de la canción con duración mas baja es: '
print(txt3 + str(cancion_duracion_min))
# Se obtiene la canción con la duración mas alta
cancion_duracion_max = songs.loc[songs['duration_ms'].idxmax(), 'track']
txt4 = 'El nombre de la canción con duración mas alta es: '
print(txt4 + str(cancion_duracion_max))
````

4.(5 %) ¿Cuál es la clave musical en la cual están compuestas menos canciones?

````python
# Para este inciso se usará la función value_counts()
# Que permite ir contando las incidencias de un evento en cada columna
# En este caso se contarán las incidencias en la columna 'key' y de la
# Lista obtenida se obtendrá la key menos común
# Se obtienen las ocurrencias de key
frecuencias_key = songs['key'].value_counts()
# De la lista de frecuencias se obtiene la que tuvo menos incidencias
key_menos_comun = frecuencias_key.idxmin()
print('la clave musical menos común es: ' + str(key_menos_comun))
````

5.(5 %) ¿Cuál es el top 10 de canciones más reproducidas de esta lista?

````python
# Para resolver este inciso se usará la función de pandas denominada
# Como nlargest(),que es una función que recibe como primer parámetro
# La cantidad de datos que se mostraran y como segundo parámetro
# La etiqueta de la columna en la cual se aplicará la función
# Se obtienen las 10 canciones mas reproducidas de la columna
# Se eliminan los duplicados en las canciones
canciones_sin_repetir = songs.drop_duplicates(subset='track', keep='first')
# Se filtra el resultado del top de canciones para imprimir en pantalla
# Únicamente el nombre de las canciones
top_10 = canciones_sin_repetir.nlargest(10, 'streams')
# top_10 = songs.nlargest(10, 'streams')
txt7 = 'La lista con las 10 canciones mas reproducidas es: \n \n'
top_10 = top_10[['track', 'streams']].to_string(index=False)
print(txt7 + str(top_10))
````
